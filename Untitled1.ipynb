```python
# Homework: Write a program that behaves like an ATM using classes, encapsulate important data and store balance amount.

class atm():
    def __init__(self,name,acc,pin,balance=0):
        self.name=name
        self.acc=acc
        self.pin=pin
        self.balance=balance
    def detail(self):
        print(f'\nAccount Owner Name: {self.name}\nAccount Number: {self.acc}\nAccount PIN: ####\nAccount Balance: {self.balance}')
    def dep(self,depo):
        self.balance=self.balance+depo
        print(f"\nCurrent account balance: Rs.{self.balance}\n")
    def wit(self,witt):
        if witt>self.balance:
            print(f"Insufficient Funds, the amount of money you have is Rs.{self.balance}")
        else:
            self.balance=self.balance-witt
            print(f"You have withdrawn: Rs.{witt}\nCurrent Balance: Rs.{self.balance}")
    def bal(self):
        print(f"Available Funds: Rs.{self.balance}")
    def pinchange(self,oldpin,newpin=0):
        if oldpin!=self.pin:
            print("Invalid PIN.")
        else:
            newpin=int(input("Enter your new PIN:"))
            self.pin=newpin
            print("PIN has been updated.")
        
    def trans(self):
        print("""\nList of Options:
              \n1. Account Details
              \n2. Check Balance
              \n3. Deposit
              \n4. Withdraw
              \n5. Change PIN
              \n6. Exit""")
        while True:
            try:
                opt=int(input("Choose an option:"))
            except:
                print("Error: Enter a correct option.\n")
                continue
            else:
                if opt==1:
                    atm.detail(self)
                elif opt==2:
                    atm.bal(self)
                elif opt==3:
                    depo=int(input("Enter Deposit Amount:"))
                    atm.dep(self,depo)
                elif opt==4:
                    witt=int(input("Enter Withdraw Amount:"))
                    atm.wit(self,witt)
                elif opt==5:
                    oldpin=int(input("Enter your current PIN:"))
                    atm.pinchange(self,oldpin)
                elif opt==6:
                    print("Thank you for using this ATM.")
                    break
name=input("Enter your name:")
acc=int(input("Enter your account number:"))
pin=int(input("Enter your PIN:"))
value=atm(name,acc,pin)
value.trans()
```

```python
#ATM using python

class ATM():
    def __init__(self, name, pin, balance):
        self.name = name
        self.pin = pin
        self.balance = balance
       
            
            
            
            
    def deposit(self, added):
        self.balance += added
        print(f'Money Deposited: Rs.{added}. Current Balance: Rs.{self.balance}')
    
    def withdraw(self, sub):
        try:
            if sub <= self.balance:  
                self.balance -= sub
                print(f'Money withdrawn: Rs.{sub}. Current Balance: Rs.{self.balance}')
            else:
                print("Insufficient Credits.")
        except ValueError:                
            print("Invalid input.")
        
    def exit(self):
        print('************ Transaction Terminated ************')
        
    def money(self):
        return self.balance
    
    
def main():
    name = input('Enter Account Name: ')
    pin = input('Enter PIN: ')
    balance = int(input('Enter current balance: '))
    account = ATM(name, pin, balance)
    
    
    
    if balance<0:
        print('Please enter correct value.')
    else:
        while True:
            print('Account Menu - ')
            print('1. Deposit Money')
            print('2. Withdraw Money')
            print('3. Check Balance')
            print('4. Exit')

            try:
                option = int(input('Choose Option 1/2/3/4: '))
                if option >= 1 and option <= 4:
                    if option == 1:
                        added = int(input('Enter money to be deposited: '))
                        account.deposit(added)
                    elif option == 2:
                        sub = int(input('Enter money to be withdrawn: '))
                        account.withdraw(sub)
                    elif option == 3:
                        print('Current Balance: Rs.', account.money())
                    elif option == 4:
                        account.exit()
                        break
                else:
                    print('Please enter a valid option.')
            except ValueError:
                print('Enter correct value.')
        

main()
```
