```python
data = pd.read_csv("churn.csv")


column_to_analyze = "Gender"


category_counts = data[column_to_analyze].value_counts()


total_entries = category_counts.sum()


category_percentages = (category_counts / total_entries) * 100


plt.bar(category_percentages.index, category_percentages.values)
plt.xlabel(column_to_analyze)
plt.ylabel("Percentage")
plt.title(f"{column_to_analyze} Distribution")
plt.xticks(rotation=45, ha="right")  
plt.ylim(0, 100)  
plt.tight_layout()  
plt.show()
```
